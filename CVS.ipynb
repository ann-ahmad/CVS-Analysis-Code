```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL Data
url = "https://docs.google.com/spreadsheets/d/1Ov0pPyGl6ps_es5cgEV5Y0vMCeFjFX-lUvVO9vgGEWs/export?format=csv"

# Load dataset and preprocess
def preprocess_data(url):
    df = pd.read_csv(url)
    # Drop timestamp if exists
    if 'Timestamp' in df.columns:
        df = df.drop(columns=['Timestamp'])

    # Mapping Gejala dan Intensitas
    frequency_map = {"Tidak pernah": 0, "Terkadang": 1, "Sering": 2}
    intensity_map = {"": 0, "Sedang": 1, "Berat": 2}
    for i in range(1, 17):
        df[f'Gejala {i}'] = df[f'Gejala {i}'].fillna("Tidak pernah").map(frequency_map)
        df[f'Intensitas {i}'] = df[f'Intensitas {i}'].fillna("").map(intensity_map)
        df[f'Skor Gejala {i}'] = df[f'Gejala {i}'] * df[f'Intensitas {i}']

    # Hitung skor total dan tentukan status CVS
    df['Skor Total CVS'] = df[[f'Skor Gejala {i}' for i in range(1, 17)]].sum(axis=1)
    df['CVS_Status'] = df['Skor Total CVS'].apply(lambda x: 1 if x >= 6 else 0)
    return df

# Load and preprocess the data
df = preprocess_data(url)

# 1. Gejala yang Paling Sering Muncul
def most_frequent_symptoms(df):
    gejala_counts = pd.DataFrame({
        'Gejala': [f'Gejala {i}' for i in range(1, 17)],
        'Frekuensi': [df[f'Gejala {i}'].sum() for i in range(1, 17)]
    }).sort_values(by='Frekuensi', ascending=False)

    print("Gejala yang paling sering muncul:")
    print(gejala_counts)

    # Visualisasi
    plt.figure(figsize=(10, 6))
    sns.barplot(data=gejala_counts, x='Frekuensi', y='Gejala', palette='viridis')
    plt.title("Gejala yang Paling Sering Muncul", fontsize=14)
    plt.xlabel("Frekuensi")
    plt.ylabel("Gejala")
    plt.tight_layout()
    plt.show()

most_frequent_symptoms(df)

# 2. Perbandingan Gejala dan Intensitas
def symptom_intensity_comparison(df):
    gejala_intensitas = pd.DataFrame({
        'Gejala': [f'Gejala {i}' for i in range(1, 17)],
        'Total Gejala': [df[f'Gejala {i}'].sum() for i in range(1, 17)],
        'Total Intensitas': [df[f'Intensitas {i}'].sum() for i in range(1, 17)]
    })
    print("Perbandingan antara gejala dan intensitas:")
    print(gejala_intensitas)

    # Visualisasi
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=gejala_intensitas, x='Total Gejala', y='Total Intensitas', hue='Gejala', palette='tab20', s=100)
    plt.title("Perbandingan Gejala dan Intensitas", fontsize=14)
    plt.xlabel("Total Gejala")
    plt.ylabel("Total Intensitas")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

symptom_intensity_comparison(df)

# 3. Korelasi Antar Gejala
def symptom_correlation(df):
    # Subset hanya kolom gejala
    gejala_columns = [f'Gejala {i}' for i in range(1, 17)]
    correlation_matrix = df[gejala_columns].corr()
    print("Korelasi antar gejala:")
    print(correlation_matrix)

    # Visualisasi Heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
    plt.title("Korelasi Antar Gejala", fontsize=14)
    plt.tight_layout()
    plt.show()

symptom_correlation(df)

# 4. Data dengan Skor CVS yang Jauh dari Rata-rata
def outlier_analysis(df):
    mean_score = df['Skor Total CVS'].mean()
    std_dev = df['Skor Total CVS'].std()
    upper_bound = mean_score + 2 * std_dev
    lower_bound = mean_score - 2 * std_dev

    outliers = df[(df['Skor Total CVS'] > upper_bound) | (df['Skor Total CVS'] < lower_bound)]
    print(f"Rata-rata Skor CVS: {mean_score:.2f}, Standar Deviasi: {std_dev:.2f}")
    print("Data outlier dengan skor jauh dari rata-rata:")
    print(outliers[['Skor Total CVS', 'CVS_Status']])

    # Visualisasi
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Skor Total CVS', color='skyblue')
    plt.axvline(upper_bound, color='red', linestyle='--', label='Upper Bound')
    plt.axvline(lower_bound, color='green', linestyle='--', label='Lower Bound')
    plt.title("Outlier Analysis of CVS Scores", fontsize=14)
    plt.legend()
    plt.tight_layout()
    plt.show()

outlier_analysis(df)

# 5. Distribusi Skor Total CVS
def distribution_cvs_score(df):
    # Statistik Deskriptif
    mean_score = df['Skor Total CVS'].mean()
    median_score = df['Skor Total CVS'].median()
    mode_score = df['Skor Total CVS'].mode()[0]
    std_dev = df['Skor Total CVS'].std()

    print("Distribusi Skor Total CVS:")
    print(f"Rata-rata: {mean_score:.2f}")
    print(f"Median: {median_score}")
    print(f"Modus: {mode_score}")
    print(f"Standar Deviasi: {std_dev:.2f}")

    # Visualisasi Distribusi
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Skor Total CVS'], kde=True, bins=20, color='blue')
    plt.axvline(mean_score, color='red', linestyle='--', label=f'Rata-rata: {mean_score:.2f}')
    plt.axvline(median_score, color='green', linestyle='--', label=f'Median: {median_score}')
    plt.title("Distribusi Skor Total CVS", fontsize=14)
    plt.xlabel("Skor Total CVS")
    plt.ylabel("Frekuensi")
    plt.legend()
    plt.tight_layout()
    plt.show()

# Panggil fungsi baru untuk menampilkan distribusi
distribution_cvs_score(df)
```

**ONE HOT ENCODING OUTPUT**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def preprocess_data(url):
    # Baca data
    df = pd.read_csv(url)
    print("\nDimensi data awal:", df.shape)

    # Drop timestamp
    df = df.drop(columns=['Timestamp'])

    # Mapping Gejala dan Intensitas
    frequency_map = {"Tidak pernah": 0, "Terkadang": 1, "Sering": 2}
    intensity_map = {"": 0, "Sedang": 1, "Berat": 2}

    # Proses gejala dan intensitas
    for i in range(1, 17):
        df[f'Gejala {i}'] = df[f'Gejala {i}'].fillna("Tidak pernah").map(frequency_map)
        df[f'Intensitas {i}'] = df[f'Intensitas {i}'].fillna("").map(intensity_map)
        df[f'Skor Gejala {i}'] = df[f'Gejala {i}'] * df[f'Intensitas {i}']

    # Hitung skor total dan tentukan status CVS
    df['Skor Total CVS'] = df[[f'Skor Gejala {i}' for i in range(1, 17)]].sum(axis=1)
    df['CVS_Status'] = df['Skor Total CVS'].apply(lambda x: 1 if x >= 6 else 0)

    # Proses untuk modeling
    columns_to_exclude = (
        [f'Gejala {i}' for i in range(1, 17)] +
        [f'Intensitas {i}' for i in range(1, 17)] +
        [f'Skor Gejala {i}' for i in range(1, 17)] +
        ['Skor Total CVS']
    )
    df = df.drop(columns=columns_to_exclude)

    # Simpan data sebelum encoding
    print("\nContoh data sebelum one-hot encoding:")
    print(df.head())
    print("\nJumlah kolom sebelum encoding:", df.shape[1])

    # Tampilkan informasi tipe data
    print("\nTipe data setiap kolom:")
    print(df.dtypes)

    # Daftar kolom kategorikal
    categorical_columns = ['Usia', 'Jenis Kelamin', 'Gangguan mata', 'Alat bantu',
                         'Aktivitas utama', 'Bidang', 'Metode aktivitas', 'Pemakaian gadget',
                         'Jenis gadget', 'Waktu penggunaan', 'Tujuan Penggunaan', 'Jarak gadget',
                         'Posisi', 'Pencahayaan', 'Perlindungan mata', 'Kecerahan layar']

    # One-hot encoding
    df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=False)

    print("\nContoh data setelah one-hot encoding:")
    print(df_encoded.head())
    print("\nJumlah kolom setelah encoding:", df_encoded.shape[1])

    # Tampilkan distribusi kategori untuk beberapa kolom contoh
    print("\nDistribusi kategori untuk beberapa kolom:")
    for col in categorical_columns[:5]:  # Ambil 5 kolom pertama sebagai contoh
        print(f"\nDistribusi {col}:")
        print(df[col].value_counts())

    return df_encoded

# URL data
url = "https://docs.google.com/spreadsheets/d/1Ov0pPyGl6ps_es5cgEV5Y0vMCeFjFX-lUvVO9vgGEWs/export?format=csv"

# Jalankan preprocessing
encoded_data = preprocess_data(url)

# Visualisasi perbandingan jumlah fitur
def plot_feature_comparison(original_cols, encoded_cols):
    plt.figure(figsize=(10, 6))
    plt.bar(['Original Features', 'Encoded Features'], [original_cols, encoded_cols])
    plt.title('Perbandingan Jumlah Fitur Sebelum dan Sesudah One-Hot Encoding')
    plt.ylabel('Jumlah Fitur')
    plt.show()

# Plot perbandingan
plot_feature_comparison(17, encoded_data.shape[1])  # 17 adalah jumlah kolom original
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.feature_selection import mutual_info_classif
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# URL Data
url = "https://docs.google.com/spreadsheets/d/1Ov0pPyGl6ps_es5cgEV5Y0vMCeFjFX-lUvVO9vgGEWs/export?format=csv"

def compare_smote_distribution(X, y):
    """
    Compare and visualize the distribution of classes before and after SMOTE resampling.
    """
    # Get original class distribution
    original_counts = pd.Series(y).value_counts().sort_index()

    # Apply SMOTE to get resampled data
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)

    # Get resampled class distribution
    resampled_counts = pd.Series(y_resampled).value_counts().sort_index()

    # Create comparison DataFrame
    comparison_df = pd.DataFrame({
        'Original': original_counts,
        'After SMOTE': resampled_counts
    })

    # Set up the visualization
    plt.figure(figsize=(10, 6))

    # Create grouped bar plot
    bar_width = 0.35
    index = np.arange(len(comparison_df.index))

    plt.bar(index - bar_width/2, comparison_df['Original'],
            bar_width, label='Original', color='skyblue')
    plt.bar(index + bar_width/2, comparison_df['After SMOTE'],
            bar_width, label='After SMOTE', color='lightcoral')

    # Customize the plot
    plt.xlabel('Class')
    plt.ylabel('Number of Samples')
    plt.title('Class Distribution Before and After SMOTE')
    plt.xticks(index, ['No CVS (0)', 'CVS (1)'])
    plt.legend()

    # Add value labels on top of each bar
    for i in index:
        plt.text(i - bar_width/2, comparison_df['Original'][i],
                str(comparison_df['Original'][i]),
                ha='center', va='bottom')
        plt.text(i + bar_width/2, comparison_df['After SMOTE'][i],
                str(comparison_df['After SMOTE'][i]),
                ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

    # Print numerical comparison
    print("\nClass Distribution Comparison:")
    print("\nOriginal class distribution:")
    print(original_counts)
    print("\nClass distribution after SMOTE:")
    print(resampled_counts)

def preprocess_data(url):
    df = pd.read_csv(url)
    print("Data shape:", df.shape)

    # Drop timestamp
    df = df.drop(columns=['Timestamp'])

    # Mapping Gejala dan Intensitas
    frequency_map = {"Tidak pernah": 0, "Terkadang": 1, "Sering": 2}
    intensity_map = {"": 0, "Sedang": 1, "Berat": 2}

    # Proses gejala dan intensitas
    for i in range(1, 17):
        df[f'Gejala {i}'] = df[f'Gejala {i}'].fillna("Tidak pernah").map(frequency_map)
        df[f'Intensitas {i}'] = df[f'Intensitas {i}'].fillna("").map(intensity_map)
        df[f'Skor Gejala {i}'] = df[f'Gejala {i}'] * df[f'Intensitas {i}']

    # Hitung skor total dan tentukan status CVS
    df['Skor Total CVS'] = df[[f'Skor Gejala {i}' for i in range(1, 17)]].sum(axis=1)
    df['CVS_Status'] = df['Skor Total CVS'].apply(lambda x: 1 if x >= 6 else 0)

    # Tampilkan informasi awal
    print("\nSampel data skor dan status CVS:")
    print(df[['Skor Total CVS', 'CVS_Status']].head())

    # Hitung dan tampilkan distribusi CVS
    cvs_counts = df['CVS_Status'].value_counts()
    print("\nJumlah CVS dan Tidak CVS:")
    print(cvs_counts)

    # Visualisasi distribusi CVS
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='CVS_Status')
    plt.title('Distribusi Status CVS')
    plt.show()

    # Visualisasi distribusi Skor Total
    plt.figure(figsize=(12, 6))
    sns.histplot(data=df, x='Skor Total CVS', bins=30)
    plt.axvline(x=6, color='r', linestyle='--', label='Batas CVS (6)')
    plt.title('Distribusi Skor Total CVS')
    plt.legend()
    plt.show()

    # Print sampel data lengkap
    print("\nSampel Data Lengkap:")
    print(df.head())

    # Proses untuk modeling
    columns_to_exclude = (
        [f'Gejala {i}' for i in range(1, 17)] +
        [f'Intensitas {i}' for i in range(1, 17)] +
        [f'Skor Gejala {i}' for i in range(1, 17)] +
        ['Skor Total CVS']
    )
    df = df.drop(columns=columns_to_exclude)

    # One-hot encoding untuk kolom kategori
    categorical_columns = ['Usia', 'Jenis Kelamin', 'Gangguan mata', 'Alat bantu',
                         'Aktivitas utama', 'Bidang', 'Metode aktivitas', 'Pemakaian gadget',
                         'Jenis gadget', 'Waktu penggunaan', 'Tujuan Penggunaan', 'Jarak gadget',
                         'Posisi', 'Pencahayaan', 'Perlindungan mata', 'Kecerahan layar']
    df = pd.get_dummies(df, columns=categorical_columns, drop_first=False)

    # Pisahkan fitur dan target
    X = df.drop(columns=['CVS_Status'])
    y = df['CVS_Status']

    # Convert to float32 to avoid data type issues with SMOTE
    X = X.astype('float32')

    return X, y

def perform_cross_validation(X, y, model, model_name, n_splits=5):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    scores = []
    for train_idx, val_idx in kf.split(X):
        X_train_fold = X.iloc[train_idx].values
        X_val_fold = X.iloc[val_idx].values
        y_train_fold = y.iloc[train_idx].values
        y_val_fold = y.iloc[val_idx].values

        # Apply SMOTE
        smote = SMOTE(random_state=42)
        X_train_resampled, y_train_resampled = smote.fit_resample(X_train_fold, y_train_fold)

        # Train and evaluate
        model.fit(X_train_resampled, y_train_resampled)
        score = model.score(X_val_fold, y_val_fold)
        scores.append(score)

    scores = np.array(scores)
    print(f"\n=== Cross Validation Scores for {model_name} with SMOTE ===")
    print(f"Fold Scores: {scores}")
    print(f"Mean CV Score: {scores.mean():.3f} (+/- {scores.std() * 2:.3f})")

    return scores.mean()

def train_and_evaluate_model(X, y, model, model_name):
    # Perform cross-validation first
    #cv_score = perform_cross_validation(X, y, model, model_name)

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

    # Convert to numpy arrays
    X_train_arr = X_train.values
    X_test_arr = X_test.values
    y_train_arr = y_train.values

    # Apply SMOTE
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_arr, y_train_arr)

    #CV only for training set
    #cv_score_train = perform_cross_validation(pd.DataFrame(X_train_resampled), y_train_resampled, model, model_name)

    # Train model
    model.fit(X_train_resampled, y_train_resampled)
    y_pred = model.predict(X_test_arr)

    print(f"\n=== {model_name} Test Set Evaluation (with SMOTE) ===")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    print(f"Test Set Accuracy: {accuracy_score(y_test, y_pred):.2f}")

    # Perform cross-validation
    cv_score = perform_cross_validation(X, y, model, model_name)

    return model, X_test, y_test, cv_score

def get_feature_importance(models, X, y, feature_names):
    importances = {}

    # Get importance scores for tree-based models
    for name, model in models.items():
        if name in ['DT', 'RF']:  # For Decision Tree and Random Forest
            importances[f"{name}_Importance"] = model.feature_importances_
        elif name == 'NB':  # For Naive Bayes, use Mutual Information
            mi_scores = mutual_info_classif(X, y, random_state=42)
            importances[f"{name}_Importance"] = mi_scores

    # Create DataFrame with all scores
    importance_df = pd.DataFrame({
        'Feature': feature_names,
        **importances
    })

    # Normalize scores to make them comparable
    for col in [col for col in importance_df.columns if col.endswith('_Importance')]:
        # Min-max normalization
        min_val = importance_df[col].min()
        max_val = importance_df[col].max()
        importance_df[col] = (importance_df[col] - min_val) / (max_val - min_val)

    # Add rankings
    for col in [col for col in importance_df.columns if col.endswith('_Importance')]:
        importance_df[f"{col[:2]}_Rank"] = importance_df[col].rank(ascending=False)

    # Calculate average rank
    rank_cols = [col for col in importance_df.columns if col.endswith('_Rank')]
    importance_df['Avg_Rank'] = importance_df[rank_cols].mean(axis=1)

    return importance_df.sort_values('Avg_Rank')

def plot_feature_importance(importance_df, top_n=10):
    plt.figure(figsize=(12, 6))

    # Get top N features by average rank
    top_features = importance_df.head(top_n)

    # Select importance columns
    importance_columns = [col for col in top_features.columns if col.endswith('_Importance')]

    # Create a more informative legend mapping
    method_names = {
        'DT_Importance': 'Decision Tree',
        'RF_Importance': 'Random Forest',
        'NB_Importance': 'Naive Bayes (MI)'
    }

    # Melt the DataFrame for plotting
    melted_df = pd.melt(
        top_features,
        id_vars=['Feature'],
        value_vars=importance_columns,
        var_name='Method',
        value_name='Importance'
    )

    # Update method names in melted DataFrame
    melted_df['Method'] = melted_df['Method'].map(method_names)

    # Create the plot
    sns.barplot(
        data=melted_df,
        x='Importance',
        y='Feature',
        hue='Method',
        palette='husl'
    )

    plt.title('Top 10 Most Important Features Across All Models\n(Normalized Scores)')
    plt.xlabel('Normalized Importance Score')
    plt.ylabel('Features')
    plt.tight_layout()
    plt.show()

def compare_models(cv_scores):
    plt.figure(figsize=(10, 6))
    models = list(cv_scores.keys())
    scores = list(cv_scores.values())

    sns.barplot(x=models, y=scores)
    plt.title('Cross-Validation Scores Comparison')
    plt.ylabel('Mean CV Score')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Preprocessing Data
X, y = preprocess_data(url)

# Add SMOTE comparison visualization
print("\nVisualizing class distribution before and after SMOTE:")
compare_smote_distribution(X, y)

# Dictionary to store CV scores
cv_scores = {}

# Train and evaluate models
models = {
    'DT': DecisionTreeClassifier(random_state=42),
    'RF': RandomForestClassifier(random_state=42),
    'NB': GaussianNB()
}

trained_models = {}
final_X_test = None
final_y_test = None

for name, model in models.items():
    trained_model, X_test, y_test, cv_score = train_and_evaluate_model(X, y, model, name)
    trained_models[name] = trained_model
    cv_scores[name] = cv_score
    if final_X_test is None:  # Save test set from first model
        final_X_test = X_test
        final_y_test = y_test

# Get and display feature importance
importance_df = get_feature_importance(trained_models, final_X_test, final_y_test, X.columns)
print("\n=== Top 10 Most Important Features ===")
display_columns = ['Feature']
display_columns.extend([col for col in importance_df.columns
                       if col.endswith('_Importance')])
print(importance_df[display_columns].head(10))

# Plot feature importance
plot_feature_importance(importance_df)

# Compare models
compare_models(cv_scores)
```
